```python
import pandas as pd
import matplotlib.pyplot as plt

    
```

```python
tbc=pd.read_excel('datasets\\salarios_sjc.xlsx')
#carregando planilha excel
```

```python
tbc.sample(10)
#verificando a leitura dos dados
```

```python
tbc.info()
#Verificando a quantidade de linhas, colunas e se há valores nulos ou incorretos
#index9005: sao 9005 linhas, columns9: são 9 colunas
```

```python
tbc.tail()
#para ver as ultimas linhas
```

```python
#Verificando os maiores salarios do municipio
tbc.sort_values(by= ["REMUNERAÇÃO DO MÊS"],ascending=False).head(6)
```

```python
#Verificando os menores salarios do municipio
tbc.sort_values(by= ["REMUNERAÇÃO DO MÊS"],ascending=True).head(6)
```

```python
#Verificado que ha 191 cargos diferentes no municipio, sendo o maior Professor I com 1872 
tbc.groupby('CARGO').size().sort_values(ascending=False)
```

```python
#Visualizando graficamente a quantidade de funcionarios por função (as 10 funções com mais cardos)
tbc.groupby('CARGO').size().sort_values(ascending=False).head(10).plot(kind='barh')
```

```python
#Verificando os funcionarios com maiores tempo de serviço 
tbc.sort_values(by= ["TEMPO DE SERVIÇO (ANOS)"],ascending=False).head(10)
```
